```python
from selenium import webdriver
from selenium.webdriver import ChromeOptions, DesiredCapabilities
from webdriver_setup import get_webdriver_for
from time import sleep
```

```python
caps = {}
caps['browser_api_name'] = 'Chrome39'
caps['os_api_name'] = 'WinXPSP2-C2'
caps['screen_resolution'] = '1024Ã—768'
caps['record_video'] = 'true'
caps['record_network'] = 'false'
caps['record_snapshot'] = 'false'
caps['chromeOptions'] = { "prefs" : { "profile.default_content_settings.cookies" : 2 } }
```

```python
x=0
while x < 100000:
    page = get_webdriver_for("chrome", desired_capabilities=caps)
    page.get('https://biletnikoffaward.com/fan-vote/')
    page.maximize_window()
    
    l=page.find_element(by='xpath',value="//input[@value='50452516']")
    l.click()
    b=page.find_element(by='id',value="pd-vote-button10973713")
    b.click()

    sleep(2)
    page.delete_cookie('PHPSESSID')
    x += 1
    
    page.close()
```
